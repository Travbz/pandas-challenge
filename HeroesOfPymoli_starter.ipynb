## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.

```python
purchase_data.head()
unique = purchase_data["Item Name"].value_counts().count()

average_price = round(purchase_data["Price"].mean(), 2)



number_of_purchases = purchase_data["Purchase ID"].count()

total_rev = purchase_data["Price"].sum()
purchase_analysis = pd.DataFrame({"Number of Unique Items": [unique], 
                             "Average Price": [average_price],
                             "Number of Purchases": [number_of_purchases],
                             "Total Revenue": [total_rev]})
display(purchase_analysis)

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
male = purchase_data[purchase_data["Gender"]=="Male"]
female = purchase_data[purchase_data["Gender"]== "Female"]
other = purchase_data[purchase_data["Gender"]== "Other / Non-Disclosed"]
male_count = male["SN"].value_counts().count()
female_count = female["SN"].value_counts().count()
other_count = other["SN"].value_counts().count()

percent_male = (male_count / purchase_data["Gender"].count()) *100
percent_female = (female_count / purchase_data["Gender"].count())*100
percent_other = (other_count / purchase_data["Gender"].count())*100
gender_dataframe = pd.DataFrame({" ":  ["Female" , " Male ", " Other"],
                                " Gender Count": [female_count, male_count, other_count],
                                "Percent of Gender":[percent_female, percent_male, percent_other]})
display(gender_dataframe)
```


## Purchasing Analysis (Gender)

```python

f_purchase_count = female["Price"].count()
m_purchase_count = male["Price"].count()
o_purchase_count = other["Price"].count()

male_avg_purchase = male["Price"].mean()
female_avg_purchase = female["Price"].mean()
other_avg_purchase = other["Price"].mean()

male_price = male["Price"].sum()
female_price = female["Price"].sum()
other_price = other["Price"].sum()

he_total_per = male.groupby(["SN"]).sum()
she_total_per = female.groupby(["SN"]).sum()
other_total_per = other.groupby(["SN"]).sum()

he_total_avg_per = he_total_per["Price"].mean()
she_total_avg_per = she_total_per["Price"].mean()
other_total_avg_per = other_total_per["Price"].mean()

gender_analysis = pd.DataFrame({"Gender":["Female", "Male", "Other"],
                                "Purchase Count":[f_purchase_count, m_purchase_count, o_purchase_count],
                                "Average Purchase Cost":[female_avg_purchase, male_avg_purchase, other_avg_purchase],
                                "Total Purchase Value":[female_price, male_price, other_price],
                                "Average Total Per Person":[she_total_avg_per, he_total_avg_per, other_total_avg_per]})
                                
display(gender_analysis)                                
```
